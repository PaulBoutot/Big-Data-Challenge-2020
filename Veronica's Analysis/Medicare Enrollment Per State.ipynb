```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def print_enumeration(collection):
    for num, value in enumerate(collection, start=1):
        print('{0}. {1}'.format(num, value))

```

```python
data = pd.read_csv('Medicare Enrollment Per State.csv') 
```

```python
data.head()
```

```python
medicare_enrolled = data.Medicare
state_pop = data.Population
states = data.State
```

```python
df = pd.DataFrame(data, columns = ['State', 'Medicare', 'Population'])
df['Medicare'] = df['Medicare'].astype(int)
df['Population'] = df['Population'].astype(int)
top = df['Medicare'].idxmax()
print("State with highest medicare enrollment is" , df['State'].iloc[top] )
mins = df['Medicare'].idxmin()
print("State with highest medicare enrollment is" , df['State'].iloc[mins] )
```

```python
#Percentage of State Enrolled
percent_medicare = [None] * 50

#create percentage of state enrolled in medicare percentage
for i in range(0,50):
    percent_medicare[i]=  round(((df['Medicare'].iloc[i]/df['Population'].iloc[i]) * 100),2)
    
df_new = pd.DataFrame(percent_medicare) 

df.insert(3, "Percent Enrolled", df_new, True) 

df.head()
```
