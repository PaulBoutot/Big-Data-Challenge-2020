```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

```python
#download the page and parse in soup
page = requests.get("https://www.healthecareers.com/search-jobs/?catid=&ps=100")
link = "https://www.healthecareers.com/search-jobs/?catid=&ps=100"
#Get HTML Content related to Chequing Account Names
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
#Find array of all the job locations for the first page of jobs
ok = soup.find_all(class_="job-location")
stringo = str(ok)

#find location in HTML
pattern = "id=\"job-results-location\">(.*?)</span>"
substring = re.findall(pattern, stringo)

state = [None] * len(substring) #Store all the states
city = [None] * len(substring) #Store all the cities

j = 0

#split by comma, put into two separate arrays -> one containing City and the other containing State
for stringy in substring:
    temp_strings = stringy.split(", ")
    if len(temp_strings) == 2:
        #print(temp_strings[1])
        city[j]=temp_strings[0]
        state[j]=temp_strings[1]
        #print(temp_strings) 
    j=j+1

    
df_new_s = pd.DataFrame(city)
df_new = pd.DataFrame(state) 
df = pd.DataFrame(df_new_s, columns = ['Cityy'])
df.insert(1, "States", df_new, True) 
df.insert(0, "City", df_new_s, True)
del df['Cityy']
df.head()
```
